```python
file = '../gitsandbox/Household_income.csv'

Household_income_df = pd.read_csv(file)

Household_income_df.head()
```

```python
annual_house_income = Household_income_df.groupby('Household Income')

count_annual_house_income = annual_house_income['Household Income'].count()

count_annual_house_income
```

```python
count_chart = count_annual_house_income.plot(kind='bar', figsize=(20,28))

count_chart.set_xlabel("Household Income Bucket")
count_chart.set_ylabel("Year")

```
