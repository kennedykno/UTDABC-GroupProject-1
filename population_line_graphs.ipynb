```python
# import neccessary and possibly neccessary tools
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
```

```python
# establish lists
year = [1980, 1990, 2000, 2010, 2020, 2022, 2030]
austin_metro_pop_growth = [383000, 569000, 911000, 1377000, 2053000, 2176000, 2453000]
```

```python
# plot Austin stand alone metro population growth

plt.plot(year, austin_metro_pop_growth, color='green', marker='o')
plt.title('Austin Metro Population Growth', fontsize=14)
plt.xlabel('year', fontsize=14)
plt.ylabel('Metro Growth (Hund thous)', fontsize=14)
plt.grid(True)
plt.savefig("Data_Visuals.png", format="png")
plt.show()
```

```python
# establish lists

year = [1980, 1990, 2000, 2010, 2020, 2022, 2030]
houston_metro_pop_growth = [2424000, 2922000, 3847000, 4976000, 6371000, 6603000, 7254000]


# plot Houston stand alone metro population growth
plt.plot(year, houston_metro_pop_growth, color='red', marker='o')
plt.title('Houston Metro Population Growth', fontsize=14)
plt.xlabel('year', fontsize=14)
plt.ylabel('Metro Growth (Mils)', fontsize=14)
plt.grid(True)
plt.savefig("houston_metro_pop_growth", format="png")
plt.show()
```

```python
# establish lists
year = [1980, 1990, 2000, 2010, 2020, 2022, 2030]
san_antonio_metro_pop_growth = [949000, 1134000, 1337000, 1771000, 2320000, 2413000, 2661000]


# plot Houston stand alone metro population growth
plt.plot(year, san_antonio_metro_pop_growth, color='blue', marker='o')
plt.title('San Antonio Metro Population Growth', fontsize=14)
plt.xlabel('year', fontsize=14)
plt.ylabel('Metro Growth (Mils)', fontsize=14)
plt.grid(True)
plt.savefig("San_Antonio_metro_pop_growth", format="png")
plt.show()
```

```python
# list
year = [1980, 1990, 2000, 2010, 2020, 2022, 2030]

# merge three graphs
plt.plot(year, san_antonio_metro_pop_growth, color='blue', marker='o')
plt.plot(year, houston_metro_pop_growth, color='red', marker='o')
plt.plot(year, austin_metro_pop_growth, color='green', marker='o')
plt.title('Metro Population Growth', fontsize=14)
plt.xlabel('year', fontsize=14)
plt.ylabel('Metro Growth (Mils)', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

```python
city_1 = "Austin"
city_2 = "Houston"
city_3 = "San Antonio" 
```
