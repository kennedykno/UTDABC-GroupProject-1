```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# File to Load (Remember to Change These)
austin = "Austin - Project.csv"
houston = "Houston - Project.csv"
san_an = "San Antonio - Project.csv"

# Read School and Student Data File and store into Pandas DataFrames
austin_pd = pd.read_csv(austin)
houston_pd = pd.read_csv(houston)
san_an_pd = pd.read_csv(san_an)
```

```python
austin_pd = austin_pd.dropna(how='any')
austin_pd
```

```python
test_pd = austin_pd.loc[(austin_pd["ID"] >= 85) & (austin_pd["ID"] <=97)]
austin_1_mean = test_pd["1 - Bedroom Austin"].mean()
houston_1_mean = test_pd["1 - Bedroom Houston"].mean()
sana_1_mean = test_pd["1 - Bedroom San Antonio"].mean()
austin_2_mean = test_pd["2 - Bedroom Austin"].mean()
houston_2_mean = test_pd["2 - Bedroom Houston"].mean()
sana_2_mean = test_pd["2 - Bedroom San Antonio"].mean()
austin_3_mean = test_pd["3 - Bedroom Austin"].mean()
houston_3_mean = test_pd["3 - Bedroom Houston"].mean()
sana_3_mean = test_pd["3 - Bedroom San Antonio"].mean()

bed_one = [austin_1_mean, houston_1_mean, sana_1_mean]
bed_two = [austin_2_mean, houston_2_mean, sana_2_mean]
bed_three = [austin_3_mean, houston_3_mean, sana_3_mean]

bed_one
```

```python
time = ["Nov'14","Dec'16","Dec'18","Dec'20","Dec'22"]
X_axis_dp = austin_pd["Date"]
Y_axis_dp = austin_pd["1 - Bedroom Austin"]
Y_axis2_dp = austin_pd["2 - Bedroom Austin"]
Y_axis3_dp = austin_pd["3 - Bedroom Austin"]
```

```python
Y_haxis_dp = austin_pd["1 - Bedroom Houston"]
Y_haxis2_dp = austin_pd["2 - Bedroom Houston"]
Y_haxis3_dp = austin_pd["3 - Bedroom Houston"]
```

```python
Y_saxis_dp = austin_pd["1 - Bedroom San Antonio"]
Y_saxis2_dp = austin_pd["2 - Bedroom San Antonio"]
Y_saxis3_dp = austin_pd["3 - Bedroom San Antonio"]
Y_saxis3_dp
```

```python
Austin_one, = plt.plot(X_axis_dp,Y_axis_dp, color="green", label = "1 Bedroom Austin")
Houston_one, = plt.plot(X_axis_dp, Y_haxis_dp, color = "blue", label = "1 Bedroom Houston")
San_antonio_one, = plt.plot(X_axis_dp, Y_saxis_dp, color = "orange", label = "1 Bedroom San Antonio")
plt.legend(handles=[Austin_one, Houston_one, San_antonio_one], loc="best")
plt.title("Median Rent for a 1 - Bedroom House")
plt.xlabel("Nov'14 - Dec'22")
plt.ylabel("Rent in $")
plt.savefig("Linegraph_1Bedroom.png")
plt.show()
```

```python
x = np.arange(3)
width = 0.2
  
# plot data in grouped manner of bar type
plt.bar(x-0.2, bed_one, width, color='cyan')
plt.bar(x, bed_two, width, color='orange')
plt.bar(x+0.2, bed_three, width, color='green')
plt.xticks(x, ['Austin', 'Houston', 'San Antonio'])
plt.xlabel("City")
plt.ylabel("Average Rent ($)")
plt.legend(["1 - Bedroom", "2 - Bedroom", "3 - Bedroom"])
plt.title("Average Rent for 2022 by City")
plt.savefig("Bargraph_Average_Rent_by_City.png")
plt.show()
```
