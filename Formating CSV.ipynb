```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# File to Load (Remember to Change These)
one_bed = "1 bedroom data (1).csv"
two_bed = "2 bedroom data.csv"
three_bed = "3 bedroom data.csv"

# Read School and Student Data File and store into Pandas DataFrames
one_bed_pd = pd.read_csv(one_bed)
two_bed_pd = pd.read_csv(two_bed)
three_bed_pd = pd.read_csv(three_bed)
one_bed_pd
```

```python
one_beds_pd = one_bed_pd.loc[(one_bed_pd["RegionName"] == "Houston") | (one_bed_pd["RegionName"] == "San Antonio") | (one_bed_pd["RegionName"] == "Austin") & (one_bed_pd["State"] == "TX") ]
one_beds_pd
one_beds_pd.to_csv("Correct_1_bed.csv",
                  encoding="utf-8", index=False, header=True)
```

```python
two_beds_pd = two_bed_pd.loc[(two_bed_pd["RegionName"] == "Houston") | (two_bed_pd["RegionName"] == "San Antonio") | (two_bed_pd["RegionName"] == "Austin") & (two_bed_pd["State"] == "TX") ]
two_beds_pd
two_beds_pd.to_csv("Correct_2_bed.csv",
                  encoding="utf-8", index=False, header=True)
two_beds_pd
```

```python
three_beds_pd = three_bed_pd.loc[(three_bed_pd["RegionName"] == "Houston, TX") | (three_bed_pd["RegionName"] == "San Antonio, TX") | (three_bed_pd["RegionName"] == "Austin, TX") ]
three_beds_pd
three_beds_pd.to_csv("Correct_3_bed.csv",
                  encoding="utf-8", index=False, header=True)
three_beds_pd
```
