```python
median_household_income = [75752, 53600, 53420]

x_axis = np.arange(len(median_household_income))

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Austin", "Houston", "Sanantonio"])

plt.xlim(-0.75, len(x_axis)-0.25)

plt.ylim(0, max(median_household_income)+5000)

plt.title("Median Household Income per city")
plt.xlabel("Median Household Income")
plt.ylabel("City")
```

```python
# bar graph comparing median houshold income in each city.

plt.bar(x_axis, median_household_income, color='r', alpha=0.5, align="center")

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Austin", "Houston", "San Antonio"])

plt.title("Median Household Income per city")
plt.xlabel("City")
plt.ylabel("Median Household Income")
plt.savefig("Median Household Income.png", format="png")
```
