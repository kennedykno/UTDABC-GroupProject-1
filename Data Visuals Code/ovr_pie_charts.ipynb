```python
#Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
```

```python
#Pull data file
arv_data_path = 'data/Austin_Rent_vs_Own.csv'

#Read the file
arv_data = pd.read_csv(arv_data_path)
arv_data
```

```python
arv_data = arv_data.drop(arv_data[arv_data['Year'] != 2020].index)
arv_data = arv_data.drop(arv_data[arv_data['Geography'] == 'United States'].index)
arv_data = arv_data.drop(arv_data[arv_data['Geography'] == 'Texas'].index)
arv_data.reset_index()
```

```python
#Locate and calculate Austin's percentage of housing resided in by owners
aoc = arv_data.iloc[2, 9]
austin_owner_occupied = aoc * 100
austin_renters = 100 - austin_owner_occupied
print(austin_owner_occupied)
print(austin_renters)
```

```python
# Generate a pie plot showing the distribution of homes in Austin occupied by their owners vs renters
%matplotlib notebook
import matplotlib.pyplot as plt

aoc_df = pd.DataFrame({'Austin Housing Occupants': [45.5, 54.5]},
                  index=['Owners', 'Renters'])
aoc_plot = aoc_df.plot.pie(y='Austin Housing Occupants', figsize=(5, 5), autopct="%1.1f%%")
plt.legend([])

# Save the figure
plt.savefig("Data Visuals/Austin_OvR_Pie")
```
