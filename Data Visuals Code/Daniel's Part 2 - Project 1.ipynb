```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# File to Load (Remember to Change These)
one_bed = "../CSV Files used/Correct_1_bed.csv"
two_bed = "../CSV Files used/Correct_2_bed.csv"
three_bed = "../CSV Files used/Correct_3_bed.csv"
net_profit = "../CSV Files used/Ammortization .csv"
# Read School and Student Data File and store into Pandas DataFrames
one_bed_pd = pd.read_csv(one_bed)
two_bed_pd = pd.read_csv(two_bed)
three_bed_pd = pd.read_csv(three_bed)
net_pd = pd.read_csv(net_profit)
```

```python
net_pd
```

```python
test_1_pd = one_bed_pd.loc[(one_bed_pd["ID"] >= 261) & (one_bed_pd["ID"] <=273)]
austin_1_mean = test_1_pd["Austin"].mean()
houston_1_mean = test_1_pd["Houston"].mean()
sana_1_mean = test_1_pd["San Antonio"].mean()
test_2_pd = two_bed_pd.loc[(two_bed_pd["ID"] >= 261) & (two_bed_pd["ID"] <=273)]
austin_2_mean = test_2_pd["Austin"].mean()
houston_2_mean = test_2_pd["Houston"].mean()
sana_2_mean = test_2_pd["San Antonio"].mean()
test_3_pd = three_bed_pd.loc[(three_bed_pd["ID"] >= 309) & (three_bed_pd["ID"] <=321)]
austin_3_mean = test_3_pd["Austin"].mean()
houston_3_mean = test_3_pd["Houston"].mean()
sana_3_mean = test_3_pd["San Antonio"].mean()
test_np_one = net_pd.loc[(net_pd["ID"] >= 1) & (net_pd["ID"] <2)]
np_aone = test_np_one["Austin_NP"].mean()
np_hone = test_np_one["Houston_NP"].mean()
np_sone = test_np_one["San Antonio_NP"].mean()

test_np_two = net_pd.loc[(net_pd["ID"] == 2)]
np_atwo = test_np_two["Austin_NP"].mean()
np_htwo = test_np_two["Houston_NP"].mean()
np_stwo = test_np_two["San Antonio_NP"].mean()
test_np_three = net_pd.loc[(net_pd["ID"] == 3)]
np_athree = test_np_three["Austin_NP"].mean()
np_hthree = test_np_three["Houston_NP"].mean()
np_sthree = test_np_three["San Antonio_NP"].mean()
bed_one = [austin_1_mean, houston_1_mean, sana_1_mean]
bed_two = [austin_2_mean, houston_2_mean, sana_2_mean]
bed_three = [austin_3_mean, houston_3_mean, sana_3_mean]
np_one = [np_aone, np_hone, np_sone]
np_two = [np_atwo, np_htwo, np_stwo]
np_three = [np_athree, np_hthree, np_sthree]

np_one
```

```python
X_axis_dp = one_bed_pd["Date"]
Y_axis_dp = one_bed_pd["Austin"]
X_axis2_dp = two_bed_pd["Date"]
Y_axis2_dp = two_bed_pd["Austin"]
X_axis3_dp = three_bed_pd["Date"]
Y_axis3_dp = three_bed_pd["Austin"]
```

```python

Y_haxis_dp = one_bed_pd["Houston"]

Y_haxis2_dp = two_bed_pd["Houston"]

Y_haxis3_dp = three_bed_pd["Houston"]
```

```python
Y_saxis_dp = one_bed_pd["San Antonio"]

Y_saxis2_dp = two_bed_pd["San Antonio"]

Y_saxis3_dp = three_bed_pd["San Antonio"]
```

```python
Austin_one, = plt.plot(X_axis_dp,Y_axis_dp, color="green", label = "1 Bedroom Austin")
Houston_one, = plt.plot(X_axis_dp, Y_haxis_dp, color = "blue", label = "1 Bedroom Houston")
San_antonio_one, = plt.plot(X_axis_dp, Y_saxis_dp, color = "orange", label = "1 Bedroom San Antonio")
plt.legend(handles=[Austin_one, Houston_one, San_antonio_one], loc="best")
plt.title("Median Price for a 1 - Bedroom House")
plt.xlabel("Jan'00 - Dec'22")
plt.ylabel("Rent in $")
plt.savefig("../Data Visuals/Linegraph_1Bedroom_price.png")
plt.show()
```

```python
Austin_two, = plt.plot(X_axis2_dp,Y_axis2_dp, color="green", label = "2 Bedroom Austin")
Houston_two, = plt.plot(X_axis2_dp, Y_haxis2_dp, color = "blue", label = "2 Bedroom Houston")
San_antonio_two, = plt.plot(X_axis2_dp, Y_saxis2_dp, color = "orange", label = "2 Bedroom San Antonio")
plt.legend(handles=[Austin_two, Houston_two, San_antonio_two], loc="best")
plt.title("Median Price for a 2 - Bedroom House")
plt.xlabel("Jan'00 - Dec'22")
plt.ylabel("Rent in $")
plt.savefig("../Data Visuals/Linegraph_2Bedroom_price.png")
plt.show()
```

```python
Austin_three, = plt.plot(X_axis3_dp,Y_axis3_dp, color="green", label = "3 Bedroom Austin")
Houston_three, = plt.plot(X_axis3_dp, Y_haxis3_dp, color = "blue", label = "3 Bedroom Houston")
San_antonio_three, = plt.plot(X_axis3_dp, Y_saxis3_dp, color = "orange", label = "3 Bedroom San Antonio")
plt.legend(handles=[Austin_one, Houston_one, San_antonio_one], loc="best")
plt.title("Median Price for a 3 - Bedroom House")
plt.xlabel("Jan'00 - Dec'22")
plt.ylabel("Rent in $")
plt.savefig("../Data Visuals/Linegraph_3Bedroom_price.png")
plt.show()
```

```python
x = np.arange(3)
width = 0.2
  
# plot data in grouped manner of bar type
plt.bar(x-0.2, bed_one, width, color='green')
plt.bar(x, bed_two, width, color='blue')
plt.bar(x+0.2, bed_three, width, color='orange')
plt.xticks(x, ['Austin', 'Houston', 'San Antonio'])
plt.xlabel("City")
plt.ylabel("Average Home Price ($)")
plt.legend(["1 - Bedroom", "2 - Bedroom", "3 - Bedroom"])
plt.title("Average Home Price for 2022 by City")
plt.savefig("../Data Visuals/Bargraph_Average_Price_by_City.png")
plt.show()
```
